```python
def jaccard(x,y):
  a=x.intersection(y)
  b=x.union(y)
  return len(a)/len(b)
```

```python
def soundex(word: str) -> str:
  """
  Soundex implementation (slide version):
    1) Retain first letter (uppercase).
    2) Change A,E,I,O,U,H,W,Y -> '0'
    3) Map letters to digits:
          B,F,P,V -> 1
          C,G,J,K,Q,S,X,Z -> 2
          D,T -> 3
          L -> 4
          M,N -> 5
          R -> 6
    4) Remove pairs of consecutive duplicate digits.
    5) Remove all zeros.
    6) Pad with trailing zeros and return first four characters: LDDD
  """
  if not word:
      return "0000"

  w = word.strip()
  if not w:
      return "0000"
  first = w[0].upper()
  groups = {
      'B': '1', 'F': '1', 'P': '1', 'V': '1',
      'C': '2', 'G': '2', 'J': '2', 'K': '2',
      'Q': '2', 'S': '2', 'X': '2', 'Z': '2',
      'D': '3', 'T': '3',
      'L': '4',
      'M': '5', 'N': '5',
      'R': '6'
  }
  zeros = set("AEIOUHWY")
  encoded = []
  for ch in w[1:].upper():
      if ch in zeros:
          encoded.append('0')
      else:
          encoded.append(groups.get(ch, ''))
  dedup = []
  last = None
  for d in encoded:
      if d == '' :
          continue
      if d != last:
          dedup.append(d)
          last = d
  dedup_no_zeros = [d for d in dedup if d != '0']
  code = first + "".join(dedup_no_zeros)
  code = (code + "0000")[:4]
  return code
```

```python
print(soundex("Herman"))
print(soundex("Hermann"))
```
