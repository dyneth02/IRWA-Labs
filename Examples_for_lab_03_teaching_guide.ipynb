# **Inverted Index Example**



```
An inverted index is a data structure that maps each word (term) to the list of documents that contain it.
```





```
Doc1: "cat dog mouse"

Doc2: "dog elephant"

Doc3: "mouse elephant cat"
```



# **Answer**



```
{
  "cat": [Doc1, Doc3],
  "dog": [Doc1, Doc2],
  "mouse": [Doc1, Doc3],
  "elephant": [Doc2, Doc3]
}

```






```
The word "cat" appears in Doc1 and Doc3.

The word "dog" appears in Doc1 and Doc2.

The word "mouse" appears in Doc1 and Doc3.

The word "elephant" appears in Doc2 and Doc3.
```



# **Positional Index Example**


```
A positional index stores not only which documents contain a word but also the exact positions of that word within each document.
```




```
Doc1: "cat dog mouse"

Doc2: "dog elephant"

Doc3: "mouse elephant cat"
```



# **Answer**



```
{
  "cat": {
    Doc1: [1],      # "cat" is at position 1 in Doc1
    Doc3: [3]       # "cat" is at position 3 in Doc3
  },
  "dog": {
    Doc1: [2],      # "dog" is at position 2 in Doc1
    Doc2: [1]       # "dog" is at position 1 in Doc2
  },
  "mouse": {
    Doc1: [3],      # "mouse" is at position 3 in Doc1
    Doc3: [1]       # "mouse" is at position 1 in Doc3
  },
  "elephant": {
    Doc2: [2],      # "elephant" is at position 2 in Doc2
    Doc3: [2]       # "elephant" is at position 2 in Doc3
  }
}

```
